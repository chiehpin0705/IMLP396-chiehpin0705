```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')   #ignore 紅色訊息

labels = ["Python","C++","Java","JS","C","C#"]
index = np.arange(len(labels))
ratings = [5.16,5.73,14.99,3.17,11.86,4.45]
change = [1.12,0.3,-1.69,0.29,3.41,-0.45]
```

```python
min(ratings)
```

```python
plt.subplot(211)
plt.bar(index,ratings)
# plt.bar(index,change)
plt.xticks(index,labels)    # 或者可用 plt.bar(index, ratings, tick_label=labels)
# plt.xticks(rotation=45)     # 旋轉角度
plt.ylabel("Ratings")
# plt.ylim(min(ratings)*0.5, max(ratings)*1.01)
plt.title("Programming rating")
plt.show()

plt.subplot(212)
plt.barh(index,ratings)
plt.xticks(index,labels)    # 或者可用 plt.barh(index, ratings, tick_label=labels)
# plt.xticks(rotation=45)     # 旋轉角度
plt.ylabel("Ratings")
plt.title("Programming rating")
plt.show()
```

```python
plt.barh(index,change)
plt.xticks(index,labels)     # 或者可用 plt.bar(index, change, tick_label=labels)
plt.ylabel("Change")
plt.title("Programming rating")
plt.show()
```

```python
x = np.linspace(0,10,50)
sinus = np.sin(x)
sinhs = np.sinh(x)
```

```python
fig, ax = plt.subplots()  # fig, ax = plt.subplots(11)
# fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x,sinus,'r-o')
ax2 = ax.twinx()   #aw2與ax共用
ax2.plot(x,sinhs,'g--')
plt.show()
```
