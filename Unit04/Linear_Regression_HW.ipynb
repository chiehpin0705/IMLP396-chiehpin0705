## 進階題 - 切割資料集分別做訓練與預測(訓練資料80%、測試資料20%)

```python
#匯入在sklearn.linear_model套件裡面的LinearRegression模型
#匯入在sklearn.model_selection套件裡面的train_test_split模組
#切割數據集(訓練資料80%、測試資料20%，設定random_state=20)
#畫出訓練資料集的matplotlib圖形

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20)
plt.title("Train Set")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x_train, y_train)
```

## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
# 1.only train set
regr_lin = LinearRegression()
X_train = x_train.reshape(-1, 1)
regr_lin.fit(X_train, y_train)
Y_train = regr_lin.predict(X_train)

# 2.mse
mse_train = np.sum((Y_train-y_train)**2) / len(y_train)
print("mse_train: ",mse_train)

# 3.y_train v.s. y_train_predict
plt.scatter(x_train, y_train)
plt.plot(x_train, Y_train, 'r')

# 4.y_test v.s y_test_predict
X_test = x_test.reshape(-1, 1)
Y_test = regr_lin.predict(X_test)
# plt.scatter(x_test, y_test)
# plt.plot(x_test, Y_test, 'r')
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>
使用X = 2.44897959，預測出來數值應該為 4.3025375<br>
所有測試集資料
$$\widehat{y}=xw+b=w_{1}x_{1}+b$$

```python
W = regr_lin.coef_
b = regr_lin.intercept_

xPoint = 2.44897959
yPoint = W*xPoint + b
print(xPoint, yPoint)
```

```python
# 訓練集
W = regr_lin.coef_
b = regr_lin.intercept_
print("[方式一]:\n", np.dot(X_train, W.T) + b)
print("[方式二]:\n", regr_lin.predict(X_train))
```

2. 並計算出測試階段的MSE

```python
mse = np.sum((Y_test - y_test)**2) / len(y_test)
print(mse)
```
